```python
%pylab notebook
import numpy as np
import matplotlib.pyplot as plt
```

# Low Rise

```python
N = 5
waits = (5.4046, 3.5144, 3.3547, 3.7149, 3.1731)
journeys = (4.4949, 4.4700, 4.4674, 4.4719, 4.4770)
```

```python
ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, waits, width, color='darkblue')
rects2 = ax.bar(ind + width, journeys, width, color='blue')

# add some text for labels, title and axes ticks
ax.set_ylabel('Tempo')
ax.set_title('Tempo MÃ©dio de Espera e de Viagem por Scheduler (Low-rise)')
ax.set_xticks(ind + width)
ax.set_xticklabels(('Random', 'NN', 'BNN', 'Weighted', 'Planning'))

ax.legend((rects1[0], rects2[0]), ('Espera', 'Viagem'))
savefig('low-rise-times.eps', bbox_inches='tight')
```
